# Q1. Is there any difference in the data type of variables list_ and array_list? If there is then write a code to print the data types of both the variables.

Yes. There is difference in the data type of variables list_ and array_list. list_ is type of list while array_list is of type of numpy.ndarray

```python
import numpy as np
list_ = [ '1' , '2' , '3' , '4' , '5' ]
array_list = np.array(object = list_)
print(type(list_))
print(type(array_list))
```

# Q2. Write a code to print the data type of each and every element of both the variables list_ and array_list.

```python
import numpy as np
list_ = [ '1' , '2' , '3' , '4' , '5' ]
array_list = np.array(object = list_)
print("Type of every element in list_  :")
for i in list_:
    print(type(i))
print("Type of every element in array_list  :")
for i in array_list:
    print(type(i))
```

There will be no difference in data type of the elements present in variable list_
There will be difference in data type of the elements present in array_list. Element's data type will changed from numpy.str_ to numpy.int64

```python
import numpy as np

list_ = [ '1' , '2' , '3' , '4' , '5' ]
array_list = np.array(object = list_, dtype = int)

print("Type of every element in list_  :")
for i in list_:
    print(type(i))
print("Type of every element in array_list  :")
for i in array_list:
    print(type(i))
```

Q4. Write a code to find the following characteristics of variable, num_array:
(i) shape
(ii) size

```python
import numpy as np
num_list = [ [ 1 , 2 , 3 ] , [ 4 , 5 , 6 ] ]
num_array = np.array(object = num_list)
print(f'shape is {num_array.shape}')
print(f'size is {num_array.size}')
```
