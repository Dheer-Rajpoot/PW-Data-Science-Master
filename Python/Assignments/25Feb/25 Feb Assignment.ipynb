# Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?

loc and iloc are two indexing methods in Pandas that allow you to select subsets of rows and columns from a DataFrame.

The main difference between loc and iloc is the way they are indexed:

loc is label-based indexing, which means that you can select data based on the labels or names of rows and columns. You can use the actual row and column labels to select specific data from a DataFrame. For example, df.loc[0, 'column_name'] will return the value at row 0 and column 'column_name'.

iloc is integer-based indexing, which means that you can select data based on its integer position or index. You can use the integer index values to select specific data from a DataFrame. For example, df.iloc[0, 1] will return the value at row 0 and column 1.

# Q4. Write a code to find the following statistical measurements for the above dataframe df1:
# (i) mean of each and every column present in the dataframe.
# (ii) standard deviation of column, ‘column_2’

```python
import pandas as pd
import numpy as np
columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1,2,3,4,5,6]
#Creating a dataframe:
df1 = pd.DataFrame(np.random.rand(6,6), columns = columns, index = indices)
print("df1 is:")
print(df1)
# (i) Mean of each column
mean = df1.mean()
print("\nMean of each column:\n", mean)

# (ii) Standard deviation of column 'column_2'
std = df1['column_2'].std()
print("\n Standard deviation of 'column_2':", std)
```

# Q6. What do you understand about the windows function in pandas and list the types of windows functions?

In Pandas, the rolling() function is used to create a window or a rolling window object that can be used to apply various types of window functions. A window function is applied to a fixed-size subset of data (window), which then slides over a larger dataset, applying the same operation to each window.

There are several types of window functions available in Pandas:

<b>Rolling Window Functions</b>: These functions operate on a defined window of rows and return a single value for each window. Examples of rolling window functions include rolling.mean(), rolling.sum(), and rolling.std().

<b>Expanding Window Functions</b>: These functions apply the window function to all previous rows in the DataFrame, resulting in an expanding window of values. Examples of expanding window functions include expanding.mean(), expanding.sum(), and expanding.std().

<b>Exponentially Weighted Window Functions</b>: These functions assign exponentially decreasing weights to each value in the window, with the most recent value receiving the highest weight. Examples of exponentially weighted window functions include ewm.mean(), ewm.std(), and ewm.var().

<b>Rolling and Expanding apply()</b>: These functions allow you to apply a custom function to a window of data. Examples include rolling.apply() and expanding.apply().

# Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and calculates the difference between them in days, hours, and minutes using Pandas time delta. The program should prompt the user to enter the dates and display the result.

```python
import pandas as pd

# prompt user to enter dates
date1 = input("Enter first date (YYYY-MM-DD): ")
date2 = input("Enter second date (YYYY-MM-DD): ")

# convert dates to datetime objects
date1 = pd.to_datetime(date1)
date2 = pd.to_datetime(date2)

# calculate difference between two dates. Result is timedelta object
diff = date1 - date2

# display result
print("Difference between", date1.date(), "and", date2.date())
print("Days:", diff.days)
print("Hours:", diff.seconds // 3600)
print("Minutes:", (diff.seconds // 60) % 60)
```

# Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified column to a categorical data type. The program should prompt the user to enter the file path, column name, and category order, and then display the sorted data.

```python
import pandas as pd

# https://media.githubusercontent.com/media/datablist/sample-csv-files/main/files/people/people-100.csv
file_path = input("enter csv file path")
# column name - Sex
column_name = input("enter column name")
df = pd.read_csv(file_path)
#category order  = Female,Male
category_order = input("Enter the category order (comma separated): ").split(',')
#Convert column to categorical in pandas python using categorical() function
df[column_name] = pd.Categorical(df[column_name],categories=category_order)

print(df.sort_values(by = column_name, ascending=False))
```

# Q10. Write a Python program that reads a CSV file containing sales data for different products and visualizes the data using a stacked bar chart to show the sales of each product category over time. The program should prompt the user to enter the file path and display the chart.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Prompt user for file path https://raw.githubusercontent.com/dahamdone/sales_data/main/sales_data.csv
file_path = input("Enter the file path: ")

# Read CSV file into a dataframe
df = pd.read_csv(file_path)

# Prompt user for column names
year_col = input("Enter the name of the year column: ") #Year
category_col = input("Enter the name of the product category column: ") #Product_Category
quantity_col = input("Enter the name of the order sale quantity column: ") #Order_Quantity

# Group data by year and product category, and sum the order quantities
grouped = df.groupby([year_col, category_col])[quantity_col].sum()

# Create a stacked bar chart of order quantities by year and product category
grouped.unstack().plot(kind='bar', stacked=True, figsize=(10, 6))

# Set the title and axis labels
plt.title('Order Quantities by Year and Product Category')
plt.xlabel(year_col)
plt.ylabel(quantity_col)

# Display the chart
plt.show()
```
